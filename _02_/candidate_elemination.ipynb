```python
import csv
data=[]
res=[]
with open("data.csv") as f:
    reader=csv.reader(f)
    for row in reader:
        data.append(row[:-1])
        res.append(row[-1])
n=len(data[0])
print(n)
```

```python
for i,row in enumerate(data):
    print(data[i],res[i])
```

```python
s=data[0].copy()
print("iniital specific hyp:",s)
g=[['?' for _ in range(n)] for _ in range(n)] 
print("genric hyp",g)
```

```python
for i,row in enumerate(data):
    currRes=res[i]
    currRow=row
    if currRes=='Yes':
        for j in range(n):
            if s[j]!=currRow[j]:
                s[j]='?'
                g[j][j]='?'
    else:
        for j in range(n):
            if currRow[j]!=s[j]:
                g[j][j]=s[j]
            else:
                g[j][j]='?'


        
```
