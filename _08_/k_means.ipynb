```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
red=np.random.uniform(3,1,40)
blue=np.random.uniform(7,1,40)
data=np.sort(np.concatenate((red,blue))).reshape(-1,1)
print(data)
```

```python
# APPLYING K MEANS TO DATA
kmeans=KMeans(n_clusters=2,random_state=0)
kmeans.fit(data)
```

```python
# cluster centroids
print(kmeans.cluster_centers_)
```

```python
# Elbow Method
Nc = range(1, 6)
scores = []
for k in Nc:
    model = KMeans(n_clusters=k, random_state=0)
    model.fit(data)
    scores.append(model.inertia_)
# better than score()
# Plot Elbow Curve
plt.plot(Nc, scores, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```
