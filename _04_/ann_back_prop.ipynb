```python
from math import exp
def sigmoid(x):
    return 1/(1+exp(-x))
```

```python
## input nodes
x1=0.2
x2=0.9
```

```python
# input to hidden  neurons wights
w13=0.1
w14=0.2
w23=0.3
w24=0.4
```

```python
## hidden to output neuron weights
w35=0.5
w45=0.6
```

```python
# output neuron
t=0.5
mu=1
```

```python
epoch=100
for _ in range(100):

    ##forward propagation
    a3=x1*w13+x2*w23
    y3=sigmoid(a3)

    a4=x1*w14+x2*w24
    y4=sigmoid(a4)

    a5=y3*w35+y4*w45
    y5=sigmoid(a5)

    # backward propagation

    ## error calc
    e5=(t-y5)*y5*(1-y5) ## for target neron
    e3=w35*e5*y3*(1-y3)
    e4=w45*e5*y4*(1-y4)

    ## update weights
    w35+=mu*e5*y3
    w45+=mu*e5*y4
    w13+=mu*e3*x1
    w14+=mu*e4*x1
    w23+=mu*e3*x2
    w24+=mu*e4*x2
```

```python
print("final weights")
print("w13=",w13)
print("w14=",w14)
print("w23=",w23)
print("w24=",w24)
print("w35=",w35)
print("w45=",w45)
```
