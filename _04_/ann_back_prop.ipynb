```python
import numpy as np
```

```python
inputs=np.array([
 [1,2],
 [3,4],
 [5,6]
],dtype=float)
targets=np.array([
    [7],
    [8],
    [9]
],dtype=float)
```

```python
## normalize data
inputs = inputs / np.max(inputs, axis=0)
targets = targets / 100.0
```

```python
print(inputs)
print(targets)
```

```python
def activate(x):## sigmoid
    return 1/(1+np.exp(-x))
def activate_derivative(x):
    return x*(1-x)
```

```python
num_input_features = inputs.shape[1]   # 2
num_hidden_neurons = 3
num_outputs = targets.shape[1]  
```

```python
epoch=1000
training_factor=0.2 ## eta
```

```python
## input to hidden
weights_input_to_hidden = np.random.uniform(size=(num_input_features, num_hidden_neurons))
bias_hidden = np.random.uniform(size=(1, num_hidden_neurons))
```

```python
print(weights_input_to_hidden)
print(bias_hidden)
```

```python
weights_hidden_to_output=np.random.uniform(size=(num_hidden_neurons,num_outputs))
bias_output=np.random.uniform(size=(1,num_outputs))
```

```python
print(weights_hidden_to_output)
print(bias_output)
```

```python
for _ in range(epoch):
     # ---- Forward pass ----
    hidden_input = np.dot(inputs, weights_input_to_hidden) + bias_hidden
    hidden_output = activate(hidden_input)

    final_input = np.dot(hidden_output, weights_hidden_to_output) + bias_output
    predicted_output = activate(final_input)

    # ---- Calculate error ----
    output_error = targets - predicted_output

    # ---- Backpropagation ----
    output_gradient = output_error * activate_derivative(predicted_output)

    hidden_error = np.dot(output_gradient, weights_hidden_to_output.T)
    hidden_gradient = hidden_error * activate_derivative(hidden_output)

    # ---- Update weights ----
    weights_hidden_to_output += np.dot(hidden_output.T, output_gradient) * training_factor
    weights_input_to_hidden += np.dot(inputs.T, hidden_gradient) * training_factor

    # ---- Update biases ----
    bias_output += np.sum(output_gradient, axis=0, keepdims=True) * training_factor
    bias_hidden += np.sum(hidden_gradient, axis=0, keepdims=True) * training_factor
```
