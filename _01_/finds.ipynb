```bash
import csv
```

```bash
data=[]
with open("data.csv") as f:
    reader=csv.reader(f)
    for row in reader:
        data.append(row)
```

```bash
for i,row in enumerate(data):
    print(i,row)
```

```bash
ans=data[0][:-1]
print("initial hypothesis",ans)
```

```bash
for i in range(1,len(data)):
    hyp=data[i][:-1]
    res=data[i][-1]
    if res=='Yes':
        for j in range(len(hyp)):
            if ans[j]!=hyp[j]:
                ans[j]='?'

    print(ans)
        
            
```
