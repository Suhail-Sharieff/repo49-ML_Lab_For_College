```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import CategoricalNB
```

```python
data = pd.read_csv("data.csv", header=None)
data.columns = ["Age", "Gender", "FamilyHistory", "Diet", "LifeStyle", "Cholesterol", "HeartDisease"]
#
# Selects all columns EXCEPT the last one for X
X = data.iloc[:, :-1]

# Selects ONLY the last column for y
y = data.iloc[:, -1]
enc = OrdinalEncoder() ## encodes text to numbers   
X = enc.fit_transform(X)

model = CategoricalNB() ## makes predixtions based on categories
model.fit(X, y)

cols = ["Age", "Gender", "FamilyHistory", "Diet", "LifeStyle", "Cholesterol"]

while True:
    row = pd.DataFrame([[
        input("Enter Age: "),
        input("Enter Gender: "),
        input("Enter FamilyHistory: "),
        input("Enter Diet: "),
        input("Enter LifeStyle: "),
        input("Enter Cholesterol: ")
    ]], columns=cols)

    row = enc.transform(row)
    pred = model.predict(row)[0]
    prob = model.predict_proba(row)[0]

    print("Prediction:", pred)
    print("Probability:", dict(zip(model.classes_, prob)))

    if input("Continue? (y/n): ").lower() != "y":
        break
```
